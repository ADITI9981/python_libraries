```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(x,y,color='green')
```

```python
plt.plot(x,y,'g+--')
```

```python
plt.plot(x,y,'rD')
```

```python
days =[1,2,3,4,5,6,7]
max_t =[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best",shadow=True,fontsize="small")

plt.grid()
```

```python
company=['google','amazon','tcs','infosys']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95)
```

```python
blood_sugar_man=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_woman=[67,98,89,120,133,150,84,69,79,120,112,100]

plt.xlabel("sugar range")
plt.ylabel("total no. of patients")
plt.title("blood sugar analysis")
plt.hist([blood_sugar_man,blood_sugar_woman],bins=[80,100,125,150],rwidth=0.95,color=['blue','pink'],label=['men','woman'])
plt.legend()
```
